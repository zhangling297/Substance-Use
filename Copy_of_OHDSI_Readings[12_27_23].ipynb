#Chapter 10: Defining Cohorts

## [S] 10.1-10.10

Reading Notes: It's interesting and important to note that the definition of a "cohort" can be counterrintuitive. For example, you might not consider that there is a time component to inclusion/exclusion. So, adding the fact that various conditions are chronic or recurrent, the 1-to-1 relationship between "person" and "cohort member" doesn't exist. -- As you read, consider how this definition differs from the definition of a cohort that you might have held before reading. What issue arise when determining inclusion exclusion?

Finally, how do 'concept sets' help to bypass those issues, by providing a useful shorthand for producing reproducible cohorts? -- As you'll see, defining a good cohort is difficult, as is quantifying the validity of a cohort. And so, one of the great features of OMOP is the ability to start from these shorthands, and eventually modify them as needed.
