```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
cancer = load_breast_cancer()
# in the dataset, B stands for Benign, and M stands for Malignant
```

```python
cancer.data.shape #cancer.data contains only the 30 predictors, now class variable
```

```python
cancer.data
```

```python
scaler = StandardScaler()
scaler.fit(cancer.data)
X_scaled = scaler.transform(cancer.data)
```

```python
cancer.target #cancer.target shows the class variable: 0 - Malignant, 1 - Benign
```

```python
pca = PCA(n_components=2)
# fit PCA model to beast cancer data
pca.fit(X_scaled) # fit() will get the covariance matrix and the eigenvalues

# transform data onto the first two principal components
X_pca = pca.transform(X_scaled) # transform() will get the PCs
print("Original shape: {}".format(str(X_scaled.shape)))
print("Reduced shape: {}".format(str(X_pca.shape)))
```

Note: The following scatterplot is generated by the embedded GenAI tool by the following prompt "generate a scatterplot to visualize cancer.target=0 and cancer.target=1"

```python
# prompt: generate a scatterplot to visualize cancer.target=0 and cancer.target=1

import matplotlib.pyplot as plt
colors = ['red', 'green']

for i in range(len(cancer.target)):
    plt.scatter(X_pca[i, 0], X_pca[i, 1], c=colors[cancer.target[i]])

plt.legend(loc='best', shadow=False, scatterpoints=1)
plt.title('Visualization of cancer dataset')
plt.xlabel("First principal component")
plt.ylabel("Second principal component")
```
