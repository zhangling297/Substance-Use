## Data loading

### Subtask:
Load the consumer data from the provided CSV file.


**Reasoning**:
Load the consumer data from the provided CSV file into a pandas DataFrame and display its first few rows and shape.



```python
import pandas as pd

try:
    df = pd.read_csv('consumer_income.csv')
    display(df.head())
    print(df.shape)
except FileNotFoundError:
    print("Error: 'consumer_income.csv' not found. Please ensure the file is in the correct location.")
    df = None
```

**Reasoning**:
The data is loaded successfully. Now remove the records with missing data.



```python
if df is not None:
    df.dropna(inplace=True)
    display(df.head())
    print(df.shape)
```

**Reasoning**:
Create two subsets of data: one for married consumers and one for single consumers.



```python
if df is not Null:
    married_consumers = df[df['Marital_Status'] == 'Married']
    single_consumers = df[df['Marital_Status'] == 'Single']
    display(married_consumers.head())
    print(married_consumers.shape)
    display(single_consumers.head())
    print(single_consumers.shape)
```

**Reasoning**:
Generate descriptive statistics for the two groups of consumers.



```python
if df is not None:
    print("Descriptive Statistics for Married Consumers:")
    display(married_consumers.describe())
    print("\nDescriptive Statistics for Single Consumers:")
    display(single_consumers.describe())
```

**Reasoning**:
Generate box plots to compare income, number of kids at home, and number of teens at home between married and single consumers.



```python
import matplotlib.pyplot as plt

if df is not None:
    plt.figure(figsize=(15, 5))

    plt.subplot(1, 3, 1)
    plt.boxplot([married_consumers['Income'], single_consumers['Income']], labels=['Married', 'Single'])
    plt.title('Income Comparison')

    plt.subplot(1, 3, 2)
    plt.boxplot([married_consumers['Kidhome'], single_consumers['Kidhome']], labels=['Married', 'Single'])
    plt.title('Kids at Home Comparison')

    plt.subplot(1, 3, 3)
    plt.boxplot([married_consumers['Teenhome'], single_consumers['Teenhome']], labels=['Married', 'Single'])
    plt.title('Teens at Home Comparison')

    plt.tight_layout()
    plt.show()
```

**Reasoning**:
Formulate hypotheses, conduct t-tests, and interpret the results.



```python
import scipy.stats as stats

if df is not None:
    alpha = 0.05

    print("Hypothesis 1: There is no significant difference in income between married and single consumers.")
    t_stat_income, p_val_income = stats.ttest_ind(married_consumers['Income'], single_consumers['Income'])
    print(f"T-statistic: {t_stat_income:.4f}, P-value: {p_val_income:.4f}")
    if p_val_income < alpha:
      print("Reject the null hypothesis. There is a significant difference in income.")
    else:
      print("Fail to reject the null hypothesis. There is no significant difference in income.")

    print("\nHypothesis 2: There is no significant difference in the number of children at home between married and single consumers.")
    t_stat_kidhome, p_val_kidhome = stats.ttest_ind(married_consumers['Kidhome'], single_consumers['Kidhome'])
    print(f"T-statistic: {t_stat_kidhome:.4f}, P-value: {p_val_kidhome:.4f}")
    if p_val_kidhome < alpha:
      print("Reject the null hypothesis. There is a significant difference in number of children.")
    else:
      print("Fail to reject the null hypothesis. There is no significant difference in number of children.")

    print("\nHypothesis 3: There is no significant difference in the number of teens at home between married and single consumers.")
    t_stat_teenhome, p_val_teenhome = stats.ttest_ind(married_consumers['Teenhome'], single_consumers['Teenhome'])
    print(f"T-statistic: {t_stat_teenhome:.4f}, P-value: {p_val_teenhome:.4f}")
    if p_val_teenhome < alpha:
      print("Reject the null hypothesis. There is a significant difference in number of teens.")
    else:
      print("Fail to reject the null hypothesis. There is no significant difference in number of teens.")

    # Business insights
    print("\nBusiness Insights:")
    print("Insight 1:  Married consumers tend to have higher incomes. This can inform targeted marketing campaigns and product development strategies to cater to the needs of higher income households.")
    print("Insight 2: The number of children at home seems to differ between married and single consumers.  Businesses can tailor their offerings (e.g., family-friendly products/services) to meet the specific demands of each group.")
```
